```python
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

from sklearn.cluster import KMeans
```

### Read dataset and fit the points into a KMeans model

```python
df = pd.read_csv('datasets/clusters.csv')
points = df.values

model = KMeans(n_clusters=3)
model.fit(points)

xs = points[:,0]
ys = points[:,1]

## Predict the values
labels = model.predict(points)
```

### Each point will get a label, which is the index of the cluster they are part of

```python
labels
```

```python
plt.scatter(xs, ys, c=labels)
plt.show()
```

### Get cluster centers (best fit centroids that minimized the cluster squared error)

```python
centroids = model.cluster_centers_
centroids
```

```python
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
```

### Plot the clusters along with their centroid

```python
plt.scatter(xs, ys, c=labels)
plt.scatter(centroids_x, centroids_y, marker='X', s=200)
plt.show()
```
